```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

<b>Load the data file</b>

```python
with open("./data/ex1data2.txt") as f:
    file = f.readlines()
file
```

<b> create arrays from the data so that we can create a pandas dataframe for visualization purposes</b>

```python
area = np.array([])
bhk = np.array([])
price = np.array([])

for line in file:
    area = np.append(area, float(line.split(",")[0]))
    bhk = np.append(bhk, int(line.split(",")[1]))
    price = np.append(price, float(line.split(",")[2][:-1]))

print(area)
print(bhk)
print(price)
```

<b>Create Pandas dataframe</b>

```python
df = pd.DataFrame({
    "Area": area,
    "BHK": bhk,
    "Price": price
})
df
```

<b> Since this data is just random numbers there's no point to remove outliers based on the data frame, we can try to see the mean and standard deviation and see if we need ot move any point although the dataset is too low so probably removing point won't be a good idea</b>

```python
df1 = df.copy()
df1["Price_per_sqft"] = df1["Price"]/df1["Area"]
sns.kdeplot(df1.Price_per_sqft)
```

<b>Since the distribution looks fine I will just add a new column named x0 to match with theeta0 so that the calculation is fine, we also need to drop unwanted columns like price and price per sqft</b>

```python
df1.insert(0,"x0", 1)
df1
```

```python
df2 = df1.drop(["Price", "Price_per_sqft"], axis=1)
df2
```

<b>Now we can convert this to numpy array for our calculations</b>

```python
def hyp(df, theeta):
    xArray = df.to_numpy()
    return xArray.dot(theeta)
```

```python
def cost(df, theeta, df_y, m):
    yArray = df_y.to_numpy()
    return (np.sum(np.square(hyp(df, theeta) - yArray)))/(0.5*m)
```

```python
def grad(m, theeta, df_y, df):
    y = df_y.to_numpy()
    x = df.to_numpy()
    sub = hyp(df, theeta) - y
    div = sub/m
    mult = div.dot(x)
    return mult
```

```python
df3 = df2.copy()
df3.loc[:,"Area"] = (df3["Area"] - df3["Area"].mean())/(df3["Area"].std())
df3.loc[:,"BHK"] = (df3["BHK"] - df3["BHK"].mean())/(df3["BHK"].std())
df3.describe()
```

```python
df3
```

```python
# required variables
m = df3["Area"].count() #total samples
theeta = np.full(df3.shape[1], 0) #values of theeta
prev_theeta = theeta
df_y = (df1["Price"]-df1["Price"].mean())/(df1["Price"].std())
alpha = 0.001
# training the model
class GD:
    def __init__(self):
        self.i = 0
        self.cost_array = np.array([])
    
    def gd(self, theeta, prev_theeta, alpha, m, df_y, df):
        prev_cost = 5
        while True:
            theeta = prev_theeta - (alpha*grad(m, theeta, df_y, df))
            prev_theeta = theeta
            c = cost(df, theeta, df_y, m)
            self.cost_array = np.append(self.cost_array, c)
            if prev_cost - c < 100**-6:
                return theeta 
                break
            prev_cost = c
            self.i = self.i+1

# run the model
gradient_descent = GD()
parms = gradient_descent.gd(theeta, prev_theeta, alpha, m, df_y, df3)
print("cost_array - {}".format(gradient_descent.cost_array))
print("Total iterations - {}".format(gradient_descent.i))
```

```python
yAxis = gradient_descent.cost_array
xAxis = range(gradient_descent.i+1)
plt.plot(xAxis, yAxis)
plt.xlabel("Number of iteration")
plt.ylabel("Cost")
plt.show
```

```python
def predict(area, bhk):
    x = np.zeros(len(df3.columns))
    x[0] = 1
    x[1] = (area - df2["Area"].mean())/(df2["Area"].std())
    x[2] = (bhk - df2["BHK"].mean())/(df2["BHK"].std())
    
    price = x.dot(parms)
    return price*(df1["Price"].std()) + df1["Price"].mean()
```
